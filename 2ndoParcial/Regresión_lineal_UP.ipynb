## Modelo de Regresi√≥n lineal

Un **modelo de regresi√≥n lineal** es una ‚öô **funci√≥n** ‚öô capaz de relacionar ejemplos del mismo tipo de objetos y predecir su comportamiento a trav√©s de una variable.

En un modelo de regresi√≥n, la üåü **variable dependiente**, tambi√©n conocida como variable de respuesta, es aquella que se intenta **predecir** o explicar.

Las  üé¢ **variables independientes**, tambi√©n llamadas predictores o variables explicativas, son aquellas que se utilizan para predecir la variable dependiente.




La ecuaci√≥n general de la regresi√≥n lineal simple se expresa como:

 $Y = \beta_0 + \beta_1 X + \epsilon $

Donde:
- \( $Y$ \) es la variable dependiente.
- \( $X$ \) es la variable independiente.
- \( $\beta_0$ \) es el t√©rmino de intersecci√≥n o constante.
- \( $\beta_1$ \) es el coeficiente de regresi√≥n que representa la pendiente de la l√≠nea.
- \( $\epsilon$ \) es el error aleatorio.

 **Pista**: Una regresi√≥n lineal nos ayuda cuando debemos **predecir** un valor real

# Vamos a jugar con los datos

```python
#Librerias
import pandas as pd
```

```python
#Lectura de un archivo csv
data = pd.read_csv("kc_house_data.csv")
```

```python
#Vemos que columnas existen
data.columns
```

```python
#Seleccionamos solo 2 columnas para nuestro analisis
data = data[["sqft_living", "price"]]
```

#### Notaci√≥n de nuestros datos:

*   m = Numero de observaciones
*   $x_i$ = Features (variables independiente)
*   $y$ =  Variable objetivo





```python
import matplotlib.pyplot as plt

plt.scatter(data['sqft_living'], data['price'], color='blue', label='Datos')

plt.xlabel('Sqft')
plt.ylabel('Price')
plt.title('Relaci√≥n entre Pies Cuadrados y Precio de la Casa')

# Agregar una leyenda
plt.legend()

# Mostrar el gr√°fico
plt.grid(True)
plt.show()
```

#### La hipotesis
Es una funci√≥n que toma como entrada un valor x y regrea una predicci√≥n

$x_i$ -> $h(x_i)$ -> $Y_i$

```python
import matplotlib.pyplot as plt

m = 0
b = 1

x = range(-10, 11)

y = [m * xi + b for xi in x]

plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')



# Display the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(data[['sqft_living']], data['price'], test_size=0.2, random_state=42)


model = LinearRegression()

# Entrenar el modelo
model.fit(X_train, y_train)

# Coeficientes del modelo
print('Coeficiente:', model.coef_)
print('Intercepci√≥n:', model.intercept_)

# Predecir los valores para el conjunto de prueba
y_pred = model.predict(X_test)

# Graficar los datos de entrenamiento y la l√≠nea de regresi√≥n
plt.scatter(X_train, y_train, color='blue', label='Datos de Entrenamiento')
plt.scatter(X_test, y_test, color='red', label='Datos de Prueba')
plt.plot(X_test, y_pred, color='green', linewidth=3, label='L√≠nea de Regresi√≥n')
plt.xlabel('Metros Cuadrados')
plt.ylabel('Precio')
plt.title('Regresi√≥n Lineal Simple')
plt.legend()
plt.grid(True)
```
