```python
from sklearn import svm
import numpy as np

from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
```

```python
# Datos de ejemplo: [velocidad, manejo]
X = np.array([[10, 95],  # Boo: Ligero
              [20, 90],  # Toad: Ligero
              [30, 70],  # BabyLuigi: Ligero
              [50, 90],  # Coopa Tropa: Ligero
              [100, 10], # Bowser: Pesado
              [85, 40],  # Waluigi: Pesado
              [95, 20]]) # Donkey Kong: Pesado

# Etiquetas de las categorías: 0 para Ligero, 1 para Pesado
y = np.array([0, 0, 0, 0, 1, 1, 1])
```

```python
# Crear el clasificador SVM
clf = svm.SVC()

# Entrenar el clasificador con los datos existentes
clf.fit(X, y)

# Datos del nuevo personaje "Kink Boo"
nuevo_personaje = np.array([[30, 40]])

# Clasificar el nuevo personaje
etiqueta = clf.predict(nuevo_personaje)

# Imprimir la etiqueta predicha
if etiqueta == 0:
    print("El nuevo personaje 'Kink Boo' pertenece a la categoría 'Ligero'")
elif etiqueta == 1:
    print("El nuevo personaje 'Kink Boo' pertenece a la categoría 'Medio'")
else:
    print("El nuevo personaje 'Kink Boo' no se puede clasificar correctamente")
```

```python
# Crear una figura y un eje
fig, ax = plt.subplots()


# Graficar los hiperparámetros del SVM
plot_decision_regions(X, y, clf=clf, ax=ax)

# Agregar leyenda
ax.legend()

# Actualizar etiquetas de los ejes
ax.set_xlabel('Velocidad')
ax.set_ylabel('Manejo')

# Visualizar el nuevo personaje con un corazón
ax.scatter(nuevo_personaje[:, 0], nuevo_personaje[:, 1], marker='*', color='red', s=100)
```
