```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

```python
niveles = np.array([0, 5, 15, 30, 50, 60, 85, 92, 98, 99]).reshape(-1, 1)
resultados = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1])
```

```python
# 1.1 Modelo de regresión logística
modelo_logistico = LogisticRegression(solver='lbfgs')
modelo_logistico.fit(niveles, resultados)
```

```python
# 1.2 Modelo de regresión lineal (como referencia)
modelo_lineal = LinearRegression()
modelo_lineal.fit(niveles, resultados)
```

```python
# Predicciones para graficar
niveles_pred = np.arange(0, 100, 0.5).reshape(-1, 1)
proba_logistica = modelo_logistico.predict_proba(niveles_pred)[:, 1]
proba_lineal = modelo_lineal.predict(niveles_pred)
```

```python
# Gráfica de los modelos
plt.scatter(niveles, resultados, color='black', label='Datos originales')
plt.plot(niveles_pred, proba_logistica, color='blue', label='Regresión Logística')
plt.plot(niveles_pred, proba_lineal, color='green', label='Regresión Lineal')
plt.xlabel('Nivel del Entrenador')
plt.ylabel('Probabilidad de Capturar')
plt.legend()
plt.title('Modelos de Regresión')
plt.show()
```

```python
# Task 2: Probabilidad del 50% y otras probabilidades

# Encontrar el nivel para el 50% usando la fórmula del modelo logístico
nivel_50_prob = int(-modelo_logistico.intercept_[0] / modelo_logistico.coef_[0][0])

# Probabilidades para niveles específicos
niveles_especificos = np.array([10, 30, 70, 90])
for nivel in niveles_especificos:
    nivel_array = np.array([nivel]).reshape(1, -1)  # Convertir a formato 2D
    proba = modelo_logistico.predict_proba(nivel_array)[0][1]
    print(f"Nivel {nivel} --> {round(proba * 100, 2)}% probabilidad de capturar")

print(f"Nivel para 50% de probabilidad: {nivel_50_prob}")
```
